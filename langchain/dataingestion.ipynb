#### Data Ingestion- Documentloaders

https://python.langchain.com/v0.2/docs/integrations/document_loaders/

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('virat.txt')
loader
```

```python
text_doc = loader.load()
text_doc ## loaded the virat.txt as a one document
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('att.pdf')
docs = loader.load()
docs
```

```python
type(docs[0])
```

```python
from langchain_community.document_loaders import WebBaseLoader

# Initialize the WebBaseLoader with the specified web path
loader = WebBaseLoader(web_paths=["https://www.cricbuzz.com/profiles/1413/virat-kohli"])

# Load the content from the specified web page
documents = loader.load()

# Display the loaded content
for doc in documents:
    print(doc)
```
